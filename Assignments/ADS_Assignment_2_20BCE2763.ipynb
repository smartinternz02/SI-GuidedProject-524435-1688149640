# Handle the Missing values.

```python
data.isnull().sum()
```

#### we find the age column to have missing values along for which we will use the mean to impute into the data along with the embarked and embark_town and deck which are categorical columns having missing records where i have decided to drop these columns for handling the missing values

```python
data['age'].mean()
```

```python
data['age'].fillna(data['age'].mean(),inplace=True)
```

```python
data['age'].isnull().sum()
```

```python
data.isnull().sum()
```

```python
data=data.drop(['embarked','embark_town','deck'],axis=1)
```

# Find the outliers and replace the outliers

```python
sns.boxplot(data['fare'])
```

```python
perc = data.fare.quantile(0.99)
perc
```

```python
data=data[data.fare<=perc]
sns.boxplot(data.fare)
```

# label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data.sex=le.fit_transform(data.sex)
```

```python
data.alive=le.fit_transform(data.alive)
data.adult_male=le.fit_transform(data.adult_male)
data.alone=le.fit_transform(data.alone)
data.who=le.fit_transform(data.who)
```
